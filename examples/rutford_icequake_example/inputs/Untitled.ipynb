# Plot array geometry

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Import data:
stations_fname = "reftek_stations.txt"
stations_df = pd.read_csv(stations_fname)
```

```python
stations_df
```

```python
# Plot stations:
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(stations_df['Longitude'], stations_df['Latitude'])
for index, row in stations_df.iterrows():
    ax.text(row['Longitude'], row['Latitude'], row['Name'])
plt.show()
```
